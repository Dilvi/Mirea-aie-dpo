```python
# 1. Импорт библиотеки NumPy
import numpy as np

# 2. Примеры создания массивов

# 2.1. Создание массива из обычного Python-списка
array_from_list = np.array([1, 2, 3, 4, 5])
print("Массив из списка:", array_from_list)

# 2.2. Создание массива с помощью zeros
zeros_array = np.zeros((2, 3))
print("\nМассив zeros:\n", zeros_array)

# 2.3. Создание массива с помощью ones
ones_array = np.ones(5)
print("\nМассив ones:", ones_array)

# 2.4. Создание массива с помощью arange
arange_array = np.arange(0, 10)
print("\nМассив arange:", arange_array)

# 3. Демонстрация атрибутов массива(используем zeros_array как пример)

print("\nАтрибут shape (форма массива zeros_array):", zeros_array.shape)
print("Атрибут dtype (тип данных zeros_array):", zeros_array.dtype)


# 4. Примеры индексирования и срезов

# 4.1. Получение отдельного элемента
element = arange_array[3]
print("\nЭлемент arange_array[3]:", element)

# 4.2. Срез: извлечение подмассива
slice_part = arange_array[2:6]
print("Срез arange_array[2:6]:", slice_part)

# 5. Векторные операции и статистики

# 5.1. Поэлементное сложение массивов
sum_result = arange_array + np.arange(10, 20)
print("\nПоэлементное сложение:", sum_result)

# 5.2. Поэлементное умножение массива на число
mul_result = arange_array * 2
print("Поэлементное умножение:", mul_result)

# 5.3. Вычисление средней величины
mean_value = arange_array.mean()
print("\nСреднее значение массива arange_array:", mean_value)
```
